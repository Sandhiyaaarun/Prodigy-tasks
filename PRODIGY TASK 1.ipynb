# CHECK FOR MISSING VALUES

```python
missing_values = df.isnull()
missing_values
```

```python
# Summarize missing values
missing_values_count = missing_values.sum()
missing_values_count
```

# Extract relevant columns

```python
# Extract relevant columns
population_2019 = df[['Country Name', '2019']]

# Remove missing values
population_2019.dropna(inplace=True)
```

```python
population_data = df[['Country Name', '2019']]

# Drop rows with missing values
population_data.dropna(inplace=True)

# Sort data by population (optional)
population_data.sort_values(by='2019', ascending=False, inplace=True)
```

# Create the bar chart 

```python
# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(population_data['Country Name'][:10], population_data['2019'][:10])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Countries by Population in 2019')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
